```python
import os
import shutil

# Set the paths
current_folder = '/content/'
new_folder_name = 'DLNLP'
new_folder_path = os.path.join(current_folder, new_folder_name)

# Create a new folder
os.makedirs(new_folder_path, exist_ok=True)

# Move all current folders to the new folder
for item in os.listdir(current_folder):
    item_path = os.path.join(current_folder, item)
    if os.path.isdir(item_path) and item not in ['drive', new_folder_name]:
                shutil.move(item_path, os.path.join(new_folder_path, item))

# Zip the new folder
zip_file_path = '/content/drive/MyDrive/data.zip'
os.system(f"zip -r {zip_file_path} {new_folder_path}")

# Display the path to the zip file
print(f"Zipped folder saved to: {zip_file_path}")
```
